```python
!python training/ledgershield_trl_training.py \
  --output-dir artifacts/trl-openenv-hf-a10g-qwen-rich \
  --case-split all \
  --case-limit 45 \
  --train \
  --model Qwen/Qwen2.5-0.5B-Instruct \
  --max-steps 900 \
  --batch-size 1 \
  --gradient-accumulation-steps 4 \
  --max-seq-length 4096 \
  --max-new-tokens 3000 \
  --model-eval-case-limit 9 \
  --reward-eval-interval 300 \
  --reward-eval-case-limit 3 \
  --reward-eval-max-new-tokens 2200
```

```python
import json
from pathlib import Path
from IPython.display import Image, display

metrics_path = Path('artifacts/trl-openenv-hf-a10g-qwen-rich/training_metrics.json')
metrics = json.loads(metrics_path.read_text())
print(json.dumps({k: v.get('summary', {}) for k, v in metrics['evaluations'].items()}, indent=2))

for plot in sorted(Path('artifacts/trl-openenv-hf-a10g-qwen-rich/plots').glob('*.png')):
    print(plot)
    display(Image(filename=str(plot)))
```
